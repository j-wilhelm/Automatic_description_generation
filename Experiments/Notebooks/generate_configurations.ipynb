```python
def get_random_config(webshop_name, sentence_length, desc_filename_length, nr_epochs, batch_size, embeddings, extractors, 
                      architectures, optimizers, dropout, nr_nodes):
    configuration = {}
    configuration["webshop_name"] = webshop_name
    configuration["embedding_name"] = random.choice(embeddings)
    
    configuration["architecture_type"] = random.choice(architectures)
    configuration["optimizer"] = random.choice(optimizers)
    configuration["dropout"] = random.choice(dropout)
    configuration["nr_nodes"] = random.choice(nr_nodes[configuration["architecture_type"]])
    configuration["nr_epochs"] = nr_epochs[configuration["architecture_type"]]
    configuration["batch_size"] = batch_size[configuration["architecture_type"]]
    configuration["img_dim"] = (2048,)
    
    ext = random.choice(extractors)
    if configuration["architecture_type"] != "attention":
        img_folder = ext + "_GAP_last"
    else:
        img_folder = ext + "_mixed10"
    configuration["img_folder"] = img_folder
    
    return configuration
```

```python
def build_config(webshop_name, sentence_length, desc_filename_length, nr_epochs, batch_size, embedding, extractor, 
                      architecture, optimizer, dropout, nr_nodes, es, threshold):
    config = {}
    config["webshop_name"] = webshop_name
    config["sentence_length"] = sentence_length
    config["threshold"] = threshold
    config["desc_filename_length"] = 9
    config["nr_epochs"] = nr_epochs
    config["batch_size"] = batch_size
    if threshold == 10:
        ext = ".p"
    else:
        ext = ".txt"
    config["embedding_name"] = embedding.format(threshold, ext)
    config["optimizer"] = optimizer
    config["architecture_type"] = architecture
    config["dropout"] = dropout
    config["nr_nodes"] = nr_nodes
    config["img_folder"] = extractor
    config["img_dim"] = (2048,)
    config["early_stopping"] = es
    
    
    config["feat_dims"] = (8, 8, 2048)
    
    return config

def get_all_configs(webshop_name, sentence_length, desc_filename_length, nr_epochs, batch_size, embeddings, extractors, 
                      architectures, optimizers, dropout, nr_nodes, es, threshold):
    all_configs = []
    for e in embeddings:
        for a in architectures:
            for f in extractors:
                if a == "attention":
                    f = f + "_mixed10"
                else:
                    f = f + "_GAP_last"
                for d in dropout:
                    if (a == "attention") & (d == True):
                        continue
                    for n in nr_nodes[a]:
                        for o in optimizers:
                            config = build_config(webshop_name, sentence_length, desc_filename_length, nr_epochs[a], 
                                                  batch_size[a], e, f, a, o, d, n, es, threshold)
                            all_configs.append(config)
    return all_configs
                            
```

```python
def build_robust_config(webshop_name, img_folder, embedding, architecture_type, cat_embedding, batch_size, trainable_embedding, 
                       image_input=False):
    # For merge only
    config = {}
    config["webshop_name"] = webshop_name
    config["sentence_length"] = 120
    config["threshold"] = 5
    config["desc_filename_length"] = 9
    config["nr_epochs"] = 50
    config["architecture_type"] = architecture_type
    config["nr_nodes"] = 256
    config["dropout"] = False
    config["optimizer"] = "adam"
    
#     if architecture in ["merge", "parinject"]:
    config["batch_size"] = batch_size
    config["img_folder"] = img_folder
    config["embedding_name"] = embedding
    config["img_dim"] = (2048,)
    config["early_stopping"] = False    
    config["feat_dims"] = (8, 8, 2048)
    
    config["category_embedding"] = cat_embedding # whether to embed each clothing category or use a dense layer instead
    config["trainable_embedding"] = trainable_embedding # whether word embedding is trainable
    config["concat_add"] = "add"
    config["image_input"] = image_input
    return config
```
